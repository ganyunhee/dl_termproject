```python
img_link = input("Image Link:")
```

```python
import cv2
import urllib.request
import numpy as np
import matplotlib.pyplot as plt
```

```python
response = urllib.request.urlopen(img_link)
arr = np.array(bytearray(response.read()), dtype=np.uint8)
image = cv2.imdecode(arr, cv2.IMREAD_COLOR) # 'Load it as it is'

# OpenCV loads colors in BGR, matplotlib uses RGB
# --> convert image color to RGB
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = rgb_image

print('Dimensions: ', image.shape)
print(rgb_image)
```

```python
ax = plt.axes(xticks=[], yticks=[])
ax.imshow(image)
```
